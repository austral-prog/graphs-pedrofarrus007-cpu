```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import os

os.makedirs("imgs", exist_ok=True)
```

# Line Plot

```python
days = ["Mon", "Tue", "Wed", "Thu", "Fri", "Sat", "Sun"]
temperature = [20, 22, 19, 24, 26, 25, 23]

plt.figure(figsize=(8,5))

plt.plot(
    days,
    temperature,
    color='blue',
    marker='o',
    linestyle='--',
    label='Temperature'
)

plt.title("Weekly Temperature Trends")
plt.xlabel("Days")
plt.ylabel("Temperature (°C)")
plt.legend()

plt.savefig('imgs/line_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Bar chart

```python
courses = ["Math", "Physics", "Programming", "Biology", "History"]
students = [30, 25, 40, 20, 15]

plt.figure(figsize=(8,5))

plt.bar(courses, students, color='green')

plt.title("Students per Course")
plt.xlabel("Courses")
plt.ylabel("Number of Students")

plt.savefig('imgs/bar_chart.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Histogram

```python
scores = np.random.normal(70, 10, 200)

plt.figure(figsize=(8,5))

plt.hist(scores, bins=15, color='orange', edgecolor='black')

plt.title("Distribution of Exam Scores")
plt.xlabel("Scores")
plt.ylabel("Frequency")

plt.savefig('imgs/histogram.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Scatter plot

```python
hours = np.random.randint(1, 10, 50)
exam_scores = hours * 10 + np.random.randint(-10, 10, 50)

plt.figure(figsize=(8,5))

plt.scatter(hours, exam_scores, color='red')

plt.title("Study Hours vs Exam Scores")
plt.xlabel("Hours Studied")
plt.ylabel("Exam Score")

plt.savefig('imgs/scatter_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```

# Pie chart

```python
activities = ["Sleeping", "Studying", "Gaming", "Exercise", "Eating"]
hours_spent = [8, 6, 4, 2, 4]

plt.figure(figsize=(8,8))

plt.pie(
    hours_spent,
    labels=activities,
    autopct='%1.1f%%',
    startangle=90
)

plt.title("Daily Activities Distribution")

plt.savefig('imgs/custom_plot.png', dpi=300, bbox_inches='tight')

plt.show()
```
